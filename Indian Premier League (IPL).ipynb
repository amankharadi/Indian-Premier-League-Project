```python
plt.figure(figsize=(15,10))
sns.countplot('season',data=matches_df)
plt.title("Number of Matches played in each IPL season")
plt.xlabel('season')
plt.ylabel('Matches')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot('winner',data=matches_df)
plt.title("Numbers of matches won by team ")
plt.xticks(rotation=50)
plt.xlabel('Team')
plt.ylabel('Number of wins matches')
plt.show()
```
